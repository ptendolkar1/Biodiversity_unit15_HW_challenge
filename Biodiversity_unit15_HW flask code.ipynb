```python
# FOR PIE CHART 

samples_otu_maxvals_list = []  # in same sequence as sample_names

samples_df = pd.read_csv("./DataSets/belly_button_biodiversity_samples.csv", index_col=None)
# otu_id_df = pd.read_csv("./DataSets/belly_button_biodiversity_otu_id.csv", index_col=None)
# fill NaNs with 0's
samples_df = samples_df.fillna(0)  

for sample in sample_names:
    # get the 2 required columns; one is passed sample in function arg.
    id_values_df = samples_df[['otu_id', sample]].copy() #extract two columns for the sample
    values_descend_df = id_values_df.sort_values(sample, ascending=False)
    values_descend_df[sample] = values_descend_df[sample].astype(int) # convert sample column values to 'int's
    values_descend_df.columns = ['otu_id', 'sample_values'] # rename columns as per instruction
    values_descend_df = values_descend_df.reset_index(drop=True) # reset index from 0, drops previous index column
    sample_otu_maxval = values_descend_df.values.tolist()[0] # get 1st df row (max value) & convert into list [x, y]
    sample_otu_maxval.insert(0, sample) # insert sample name in its list of data as 1st item
    samples_otu_maxvals_list.append(sample_otu_maxval)
    #print(samples_otu_maxvals_list)
    
samples_otu_maxvals_list.sort(key=lambda x: x[2], reverse=True) # sorts by value i.e.by [2]; reverse for descending
samples_otu_max10 = samples_otu_maxvals_list[:10] # top 10 samples by value

otu_id_df = pd.read_csv("./DataSets/belly_button_biodiversity_otu_id.csv", index_col=None)
otu_id_list = otu_id_df.values.tolist() # convert into list [x, y]

otu_names = []
for item in samples_otu_max10:
    for otu in otu_id_list:
        if item[1] == otu[0]:
            item.insert(2, otu[1])

samples_otu_max10
```
